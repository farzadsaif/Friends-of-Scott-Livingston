```python
# Requires installation of "wordcloud" package before running script

import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
%matplotlib inline
```

Here I import contacts from these webinars:  
Februrary 19, 2019  
April 30, 2021  
June 11, 2021

```python
df = pd.read_csv("Contacts From Web Confrences.csv")
```

```python
df
```

```python
# Generate wordcloud with all descriptions
text = " ".join(df.description)
wordcloud = WordCloud(background_color = "white").generate(text)
```

```python
# Plot wordcloud

plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
# Generate default set of stopwords
# Stopwords are words that are not included in the wordcloud

stopwords = set(STOPWORDS)
```

```python
# Adding some stopwords of my own

stopwords.update(['building', 'transforming', 'cost', 'commodities', 'revitalizing',\
                  'commercialization', 'industrial', 'development', 'technologies', 'opportunities',\
                 'innovation', 'new', 'investment', 'low', 'systems', 'needs', 'battle'])
```

```python
wordcloud_descriptive = WordCloud(stopwords = stopwords, background_color = 'white').generate(text)
```

```python
fig = plt.figure()
plt.imshow(wordcloud_descriptive, interpolation = 'spline16')
plt.axis("off")
plt.show()
```

```python
# Save figure as PDF

fig.savefig('wordcloud.pdf', dpi = 1000)
```
