```python
import pandas as pd
from wordcloud import STOPWORDS
```

```python
# Import sample dataset
df = pd.read_csv("Contacts From Web Confrences.csv")
df
```

```python
# Join all descriptions so they form a block of text. Each description is separated with a space " "
text = " ".join(df.description)
word = text.split()
```

```python
# Set up stopwords: a common group of words that are not descriptive of the person
stopwords = set(STOPWORDS)
stopwords.update(['building', 'transforming', 'cost', 'commodities', 'revitalizing',\
                  'commercialization', 'industrial', 'development', 'technologies', 'opportunities',\
                 'innovation', 'new', 'investment', 'low', 'systems', 'needs', 'battle', '–'])
stopwords = list(set(stopwords) & set(word))
```

```python
stopwords
```

```python
# Remove stopwords before sorting
word = list(set(word) - set(stopwords))
```

```python
# Sort words by frequency
l = {}
for word in word:
    # make everything lowercase
    word = word.lower()
    
    # check for words in () or []
    if word.startswith('[') or word.startswith('('):
        word = word[1:]
    if word.endswith(']') or word.endswith(')'):
        word = word[:-1]
        
    # check for words that end with punctuation
    if word.endswith('.') or word.endswith(',') or word.endswith(';') or word.endswith(':'):
        word = word[:-1]
    # if word is in dictionary, increment the value
    # otherwise add the word to dictionary with value 1
    if word in l:
        l[word] += 1
    else:
        l[word] = 1
```

```python
# Done
# For this small dataset no word is used more than once. But for larger datasets this should work fine
l
```
